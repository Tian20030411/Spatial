```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

def plot_geotiff(
    tif_path,
    out_png,
    title,
    cbar_label,
    nodata_value=None,
    vmin=None,
    vmax=None
):
    with rasterio.open(tif_path) as src:
        arr = src.read(1)
        extent = [
            src.bounds.left,
            src.bounds.right,
            src.bounds.bottom,
            src.bounds.top
        ]

    if nodata_value is not None:
        arr = np.where(arr == nodata_value, np.nan, arr)

    plt.figure(figsize=(6, 8))
    im = plt.imshow(
        arr,
        extent=extent,
        origin="upper",
        vmin=vmin,
        vmax=vmax
    )
    plt.title(title, fontsize=12)
    cbar = plt.colorbar(im, fraction=0.035, pad=0.02)
    cbar.set_label(cbar_label, fontsize=10)

    plt.xlabel("Easting (m)")
    plt.ylabel("Northing (m)")
    plt.tight_layout()
    plt.show()

    plt.close()
```

```python
plot_geotiff(
    tif_path="wind_mean_aligned.tif",
    out_png="wind_mean_aligned.png",
    title="Annual Mean Wind Speed (aligned to terrain grid)",
    cbar_label="Wind speed (m/s)",
    nodata_value=-9999
)
```

```python
import os, zipfile, glob
import numpy as np
import rasterio
from rasterio.merge import merge
import matplotlib.pyplot as plt

DTM_ZIP = "TERRIAN 50 DTM.zip"  
EXTRACT_DIR = "_dtm_extract"     

OUT_SLOPE_TIF = "dtm_slope_deg.tif"
OUT_SUIT_TIF  = "slope_suitability_01.tif"
OUT_SLOPE_PNG = "dtm_slope_deg.png"
OUT_SUIT_PNG  = "slope_suitability_01.png"


if not os.path.exists(EXTRACT_DIR):
    os.makedirs(EXTRACT_DIR, exist_ok=True)
    with zipfile.ZipFile(DTM_ZIP, "r") as z:
        z.extractall(EXTRACT_DIR)

patterns = ["**/*.tif", "**/*.tiff", "**/*.asc", "**/*.bil", "**/*.img"]
rasters = []
for p in patterns:
    rasters += glob.glob(os.path.join(EXTRACT_DIR, p), recursive=True)

rasters = [r for r in rasters if os.path.isfile(r)]

if len(rasters) == 0:
    raise FileNotFoundError("grid was not found")

srcs = []
for r in rasters:
    try:
        srcs.append(rasterio.open(r))
    except Exception as e:
        

if len(srcs) == 0:
    raise RuntimeError("No grid was read.")

mosaic, out_transform = merge(srcs)  
dtm = mosaic[0].astype("float32")


meta = srcs[0].meta.copy()
meta.update({
    "height": dtm.shape[0],
    "width": dtm.shape[1],
    "transform": out_transform,
    "count": 1,
    "dtype": "float32"
})

nodata = srcs[0].nodata
if nodata is not None:
    dtm = np.where(dtm == nodata, np.nan, dtm)

dtm = np.where(dtm < -1000, np.nan, dtm)

sea_mask = np.isnan(dtm) | (dtm <= 0)

for s in srcs:
    s.close()

px = abs(out_transform.a)
py = abs(out_transform.e)

dtm_fill = np.where(np.isnan(dtm), 0.0, dtm)

dz_dy, dz_dx = np.gradient(dtm_fill, py, px)  
slope_rad = np.arctan(np.sqrt(dz_dx**2 + dz_dy**2))
slope_deg = (slope_rad * 180.0 / np.pi).astype("float32")


slope_deg = np.where(sea_mask, np.nan, slope_deg)

slope_meta = meta.copy()
slope_meta.update({"nodata": -9999})
with rasterio.open(OUT_SLOPE_TIF, "w", **slope_meta) as dst:
    dst.write(np.where(np.isnan(slope_deg), -9999, slope_deg).astype("float32"), 1)

s = slope_deg.copy()
suit = np.zeros_like(s, dtype="float32")

suit = np.where(s <= 3, 1.0, suit)
suit = np.where((s > 3) & (s <= 8), 0.7, suit)
suit = np.where((s > 8) & (s <= 15), 0.3, suit)
suit = np.where(s > 15, 0.0, suit)


suit = np.where(sea_mask, np.nan, suit)

suit_meta = meta.copy()
suit_meta.update({"nodata": -9999})
with rasterio.open(OUT_SUIT_TIF, "w", **suit_meta) as dst:
    dst.write(np.where(np.isnan(suit), -9999, suit).astype("float32"), 1)

def extent_from_transform(transform, width, height):
    left, top = transform * (0, 0)
    right, bottom = transform * (width, height)
    return (left, right, bottom, top)

extent = extent_from_transform(out_transform, dtm.shape[1], dtm.shape[0])

valid = slope_deg[~np.isnan(slope_deg)]
if len(valid) > 0:
    vmin = np.percentile(valid, 2)
    vmax = np.percentile(valid, 98)
    vmax = max(vmax, 10)
else:
    vmin, vmax = 0, 35
plt.figure(figsize=(6, 8))
plt.gca().set_facecolor("#b7d9ee")  
im1 = plt.imshow(slope_deg, extent=extent, origin="upper", vmin=vmin, vmax=vmax)
plt.title("Slope (degrees) derived from DTM", fontsize=14)
cb1 = plt.colorbar(im1, fraction=0.04, pad=0.02)
cb1.set_label("Slope ()")
plt.xlabel("Easting (m)")
plt.ylabel("Northing (m)")
plt.tight_layout()
plt.savefig(OUT_SLOPE_PNG, dpi=300, bbox_inches="tight")
plt.show()
plt.close()

plt.figure(figsize=(6, 8))
plt.gca().set_facecolor("#b7d9ee") 
im2 = plt.imshow(suit, extent=extent, origin="upper", vmin=0, vmax=1)
plt.title("Slope Suitability (01)", fontsize=14)
cb2 = plt.colorbar(im2, fraction=0.04, pad=0.02)
cb2.set_label("Suitability (01)")
plt.xlabel("Easting (m)")
plt.ylabel("Northing (m)")
plt.tight_layout()
plt.savefig(OUT_SUIT_PNG, dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```

```python
import os
import numpy as np
import rasterio
from rasterio.warp import reproject, Resampling
import matplotlib.pyplot as plt

FACTOR_TIFS = {
    "wind":      "wind_suitability_01.tif",
    "grid":      "grid_distance_suitability_01.tif",
    "slope":     "slope_suitability_01.tif",
    "road":      "road_distance_suitability_01.tif",
    "landcover": "landcover_s.tif",  
}
CONSTRAINT_TIF = "parks_suitability_01.tif"     
LCM_TIF = "LCM.tif"                            

TEMPLATE_TIF = FACTOR_TIFS["grid"]

OUT_FINAL_TIF = "final_suitability.tif"
OUT_FINAL_PNG = "final_suitability.png"
OUT_NODATA = -9999.0

LCM_EXCLUDE_CODES = {20, 21, 13, 14}

assert os.path.exists(TEMPLATE_TIF), f"Missing template: {TEMPLATE_TIF}"
with rasterio.open(TEMPLATE_TIF) as tmp:
    T_CRS = tmp.crs
    T_TRANSFORM = tmp.transform
    T_SHAPE = (tmp.height, tmp.width)
    T_PROFILE = tmp.profile.copy()

print("[OK] Template grid")
print("  CRS  :", T_CRS)
print("  shape:", T_SHAPE)
print("  res  :", (T_TRANSFORM.a, -T_TRANSFORM.e))

def align_to_template(src_path, resampling=Resampling.bilinear, dst_dtype="float32", dst_nodata=OUT_NODATA):
    
    if not os.path.exists(src_path):
        raise FileNotFoundError(f"Missing file: {src_path}")

    with rasterio.open(src_path) as src:
        src_crs = src.crs if src.crs is not None else T_CRS
        if src.crs is None:
            print(f"[WARN] {os.path.basename(src_path)} CRS=None -> assume template CRS {T_CRS}")

        src_nodata = src.nodata

        dst = np.full(T_SHAPE, dst_nodata, dtype=dst_dtype)

        reproject(
            source=rasterio.band(src, 1),
            destination=dst,
            src_transform=src.transform,
            src_crs=src_crs,
            src_nodata=src_nodata,
            dst_transform=T_TRANSFORM,
            dst_crs=T_CRS,
            dst_nodata=dst_nodata,
            resampling=resampling,
        )

    dst = dst.astype("float32")
    dst = np.where(dst == dst_nodata, np.nan, dst)
    return dst

layers = {}
for k, path in FACTOR_TIFS.items():
    resamp = Resampling.nearest if k == "landcover" else Resampling.bilinear
    layers[k] = align_to_template(path, resampling=resamp, dst_dtype="float32", dst_nodata=OUT_NODATA)
    print(f"[OK] aligned {k:<9s} -> min/max: {np.nanmin(layers[k]):.3f}/{np.nanmax(layers[k]):.3f}")

parks = align_to_template(CONSTRAINT_TIF, resampling=Resampling.nearest, dst_dtype="float32", dst_nodata=OUT_NODATA)
parks = np.where(np.isnan(parks), np.nan, np.where(parks >= 0.5, 1.0, 0.0)).astype("float32")
print(f"[OK] aligned parks -> unique: {np.unique(parks[~np.isnan(parks)])}")

lcm = align_to_template(LCM_TIF, resampling=Resampling.nearest, dst_dtype="float32", dst_nodata=OUT_NODATA)
lcm_int = np.where(np.isnan(lcm), np.nan, np.rint(lcm)).astype("float32")  

lcm_allowed = np.where(
    np.isnan(lcm_int),
    np.nan,
    np.where(np.isin(lcm_int, list(LCM_EXCLUDE_CODES)), 0.0, 1.0)
).astype("float32")
print(f"[OK] LCM hard-constraint -> unique (nan ignored): {np.unique(lcm_allowed[~np.isnan(lcm_allowed)])}")

hard_constraint = np.where(np.isnan(parks) | np.isnan(lcm_allowed), np.nan, parks * lcm_allowed).astype("float32")
print(f"[OK] hard_constraint unique (nan ignored): {np.unique(hard_constraint[~np.isnan(hard_constraint)])}")

criteria = ["wind", "grid", "slope", "road", "landcover"]

A = np.array([
    [1,   2,   5,   3,   5],
    [1/2, 1,   3,   2,   3],
    [1/5, 1/3, 1,   1/2, 2],
    [1/3, 1/2, 2,   1,   2],
    [1/5, 1/3, 1/2, 1/2, 1],
], dtype="float64")

def ahp_weights_and_cr(A):
    eigvals, eigvecs = np.linalg.eig(A)
    idx = np.argmax(eigvals.real)
    lambda_max = float(eigvals.real[idx])
    w = eigvecs[:, idx].real
    w = w / w.sum()

    n = A.shape[0]
    CI = (lambda_max - n) / (n - 1)

    RI_table = {1:0.00, 2:0.00, 3:0.58, 4:0.90, 5:1.12, 6:1.24, 7:1.32, 8:1.41, 9:1.45, 10:1.49}
    RI = RI_table.get(n, 1.49)
    CR = 0.0 if RI == 0 else CI / RI
    return w.astype("float64"), lambda_max, float(CI), float(CR)

w, lambda_max, CI, CR = ahp_weights_and_cr(A)
weights = dict(zip(criteria, w))

for c in criteria:
    print(f"{c:<10s}: {weights[c]:.4f}")
print(f"lambda_max = {lambda_max:.4f}")
print(f"CI         = {CI:.4f}")
print(f"CR         = {CR:.4f}")

for k in layers:
    layers[k] = np.clip(layers[k], 0.0, 1.0)

S = np.zeros(T_SHAPE, dtype="float32")
for k in criteria:
    S += (weights[k] * layers[k].astype("float32"))

S_final = np.where(np.isnan(hard_constraint), np.nan, S)
S_final = np.where(hard_constraint == 1.0, S_final, 0.0).astype("float32")

out_profile = T_PROFILE.copy()
out_profile.update(dtype="float32", count=1, nodata=OUT_NODATA, compress="deflate")

to_write = np.where(np.isnan(S_final), OUT_NODATA, S_final).astype("float32")
with rasterio.open(OUT_FINAL_TIF, "w", **out_profile) as dst:
    dst.write(to_write, 1)

print(f"\n[OK] Saved final suitability: {OUT_FINAL_TIF}")

plt.figure(figsize=(6, 8))
show_arr = np.where(to_write == OUT_NODATA, np.nan, to_write)
plt.imshow(show_arr, vmin=0, vmax=1)
plt.title("Final Suitability (AHP-WLC with Parks + Urban/Water hard constraints)")
plt.colorbar(label="Suitability (01)", fraction=0.035, pad=0.02)
plt.axis("off")
plt.tight_layout()
plt.savefig(OUT_FINAL_PNG, dpi=300, bbox_inches="tight")
plt.show()
print(f"[OK] Saved figure: {OUT_FINAL_PNG}")
```

```python
import os
import numpy as np
import rasterio
from rasterio.transform import xy
import geopandas as gpd
from shapely.geometry import Point
from scipy.spatial import cKDTree

FINAL_TIF = "final_suitability.tif"    
OUT_GPKG  = "new_onshore_windfarm_candidates_20.gpkg"
OUT_LAYER = "candidates_20"

N_POINTS = 20
MIN_DIST_M = 10000   

MIN_SUITABILITY = 0.70

TOPK_CANDIDATES = 200000

EXCLUDE_EXISTING = True
EXISTING_GPKG = "wind_onshore_20MW.gpkg"   
EXISTING_LAYER = None                      
EXISTING_BUFFER_M = 5000                   

OUT_NODATA = -9999.0


assert os.path.exists(FINAL_TIF), f" {FINAL_TIF}"

with rasterio.open(FINAL_TIF) as src:
    arr = src.read(1).astype("float32")
    transform = src.transform
    crs = src.crs
    nodata = src.nodata

if nodata is None:
    nodata = OUT_NODATA

arr = np.where(arr == nodata, np.nan, arr)

print("[OK] Raster loaded")
print("  CRS:", crs)
print("  shape:", arr.shape)
print("  min/max:", np.nanmin(arr), np.nanmax(arr))

valid_mask = np.isfinite(arr) & (arr >= MIN_SUITABILITY)
valid_idx = np.where(valid_mask)

if valid_idx[0].size == 0:
    raise RuntimeError(f"No pixels are satisfied MIN_SUITABILITY={MIN_SUITABILITY}")

vals = arr[valid_idx]

order = np.argsort(vals)[::-1]
if TOPK_CANDIDATES is not None and len(order) > TOPK_CANDIDATES:
    order = order[:TOPK_CANDIDATES]

rows = valid_idx[0][order]
cols = valid_idx[1][order]
scores = vals[order]

print(f"[OK] Candidate pool: {len(scores)} cells (threshold={MIN_SUITABILITY}, topK={TOPK_CANDIDATES})")

exclude_polys = None

if EXCLUDE_EXISTING:
    if not os.path.exists(EXISTING_GPKG):
    else:
        layers = gpd.list_layers(EXISTING_GPKG)
        if EXISTING_LAYER is None:
            EXISTING_LAYER = layers.name.iloc[0]
        existing = gpd.read_file(EXISTING_GPKG, layer=EXISTING_LAYER)
        if existing.crs != crs:
            existing = existing.to_crs(crs)

        exclude_polys = existing.buffer(EXISTING_BUFFER_M).unary_union
        print(f"[OK] Existing farms loaded: {len(existing)} features")
        print(f"[OK] Exclusion buffer: {EXISTING_BUFFER_M} m")

selected = []         
selected_xy = []       

for r, c, s in zip(rows, cols, scores):
    x, y = xy(transform, r, c, offset="center")

    if exclude_polys is not None:
        if exclude_polys.contains(Point(x, y)):
            continue

    if len(selected_xy) == 0:
        selected.append((x, y, float(s), int(r), int(c)))
        selected_xy.append((x, y))
    else:
        tree = cKDTree(np.array(selected_xy))
        dist, _ = tree.query(np.array([[x, y]]), k=1)
        if float(dist[0]) >= MIN_DIST_M:
            selected.append((x, y, float(s), int(r), int(c)))
            selected_xy.append((x, y))

    if len(selected) >= N_POINTS:
        break

geoms = [Point(x, y) for x, y, *_ in selected]
gdf = gpd.GeoDataFrame(
    {
        "rank": np.arange(1, len(selected) + 1),
        "suitability": [t[2] for t in selected],
        "row": [t[3] for t in selected],
        "col": [t[4] for t in selected],
        "easting": [t[0] for t in selected],
        "northing": [t[1] for t in selected],
    },
    geometry=geoms,
    crs=crs
)

if os.path.exists(OUT_GPKG):
    os.remove(OUT_GPKG)

gdf.to_file(OUT_GPKG, layer=OUT_LAYER, driver="GPKG")
print(f"[OK] Saved: {OUT_GPKG}  (layer={OUT_LAYER}, n={len(gdf)})")

print("\nTop candidates (first 5):")
print(gdf.head(5)[["rank", "suitability", "easting", "northing"]])
```

```python
import os
import numpy as np
import rasterio
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point

import matplotlib.patheffects as pe

FINAL_TIF = "final_suitability.tif"
CAND_GPKG = "new_onshore_windfarm_candidates_20.gpkg"
CAND_LAYER = "candidates_20"
EXISTING_GPKG = "wind_onshore_20MW.gpkg"

OUT_PNG = "final_suitability_candidates_map.png"

assert os.path.exists(FINAL_TIF), f"Missing {FINAL_TIF}"

with rasterio.open(FINAL_TIF) as src:
    arr = src.read(1).astype("float32")
    bounds = src.bounds
    crs_r = src.crs
    nodata = src.nodata
    extent = [bounds.left, bounds.right, bounds.bottom, bounds.top]

if nodata is None:
    nodata = -9999.0

arr = np.where(arr == nodata, np.nan, arr)

print("[OK] raster CRS:", crs_r)

def sample_raster_values(raster_path, gdf):
    with rasterio.open(raster_path) as src:
        coords = [(geom.x, geom.y) for geom in gdf.geometry]
        vals = np.array([v[0] for v in src.sample(coords)], dtype="float32")
        nd = src.nodata
        if nd is not None:
            vals = np.where(vals == nd, np.nan, vals)
        return vals

assert os.path.exists(CAND_GPKG), f"Missing {CAND_GPKG}"
cands = gpd.read_file(CAND_GPKG, layer=CAND_LAYER)

if cands.crs != crs_r:
    if cands.crs is None:
        cands = cands.set_crs(crs_r, allow_override=True)
    else:
        cands = cands.to_crs(crs_r)

assert os.path.exists(EXISTING_GPKG), f"Missing {EXISTING_GPKG}"
existing = gpd.read_file(EXISTING_GPKG)

existing = existing[existing.geometry.notnull()].copy()
existing = existing[~existing.geometry.is_empty].copy()
existing = existing[existing.is_valid].copy()

geom_types = set(existing.geometry.geom_type.unique())
if not geom_types.issubset({"Point", "MultiPoint"}):
    existing["geometry"] = existing.geometry.representative_point()

if existing.crs is None:
    xs = existing.geometry.x.to_numpy()
    ys = existing.geometry.y.to_numpy()

    if (np.nanmin(xs) > -180 and np.nanmax(xs) < 180 and
        np.nanmin(ys) > -90 and np.nanmax(ys) < 90):
        print("[WARN] existing CRS is None, coordinates look like lon/lat -> assume EPSG:4326")
        existing = existing.set_crs(epsg=4326, allow_override=True)
    else:
        print("[WARN] existing CRS is None, NOT lon/lat -> assume it is already raster CRS:", crs_r)
        existing = existing.set_crs(crs_r, allow_override=True)

if existing.crs != crs_r:
    existing = existing.to_crs(crs_r)

vals = sample_raster_values(FINAL_TIF, existing)
existing = existing[np.isfinite(vals)].copy()
print("[OK] existing kept on land (finite raster):", len(existing))

plt.figure(figsize=(7.2, 9.6))

im = plt.imshow(arr, extent=extent, origin="upper", vmin=0, vmax=1, cmap="viridis")

existing.plot(
    ax=plt.gca(),
    color="lightgrey",
    markersize=5,
    alpha=0.5,
    label=f"Existing onshore wind farms (n={len(existing)})",
    zorder=3
)
cands.plot(
    ax=plt.gca(),
    color="#E41A1C",          
    edgecolor="white",
    linewidth=1,
    markersize=50,
    label="New candidates (n=20)",
    zorder=6
)


cbar = plt.colorbar(im, fraction=0.035, pad=0.02)
cbar.set_label("Suitability (01)")

plt.title("Final suitability & selected 20 new onshore wind farm sites", fontsize=14)
plt.xlabel("Easting (m)  [EPSG:27700]")
plt.ylabel("Northing (m) [EPSG:27700]")

leg = plt.legend(loc="lower left", frameon=True, framealpha=0.9)
for h in leg.legend_handles:
    try:
        h.set_alpha(0.8)
    except Exception:
        pass

plt.tight_layout()
plt.savefig(OUT_PNG, dpi=300, bbox_inches="tight")
plt.show()

print("[OK] Saved:", OUT_PNG)
```

```python
import os
import zipfile
import numpy as np

import geopandas as gpd
import rasterio
from rasterio.features import rasterize
from rasterio.plot import plotting_extent
import matplotlib.pyplot as plt

ZIP_PATH = "National_Parks_England_-4082557758695841082.zip"
TERRAIN_TIF = "terrain.tif"
OUT_TIF = "parks_suitability_01.tif"
OUT_PNG = "parks_suitability_01.png"

PARK_VALUE = 0.0      
NONPARK_VALUE = 1.0    
NODATA = -9999.0

assert os.path.exists(ZIP_PATH), f"not found {ZIP_PATH}"
assert os.path.exists(TERRAIN_TIF), f"not found {TERRAIN_TIF}"

with zipfile.ZipFile(ZIP_PATH, "r") as z:
    shp_candidates = [
        n for n in z.namelist()
        if n.lower().endswith(".shp") and ("__macosx" not in n.lower())
    ]
    if not shp_candidates:
        raise RuntimeError("zip not found .shp ")

    shp_in_zip = sorted(shp_candidates, key=len)[0]
    print("[OK] Found shp in zip:", shp_in_zip)

zip_vfs_path = f"zip://{ZIP_PATH}!{shp_in_zip}"
gdf = gpd.read_file(zip_vfs_path)

print("[OK] loaded features:", len(gdf))
print("[OK] gdf CRS:", gdf.crs)

with rasterio.open(TERRAIN_TIF) as src:
    template_crs = src.crs
    template_transform = src.transform
    template_shape = (src.height, src.width)
    template_extent = plotting_extent(src)
    template_nodata = src.nodata
    terrain_arr = src.read(1)

print("[OK] template CRS:", template_crs)
print("[OK] template shape:", template_shape)

if gdf.crs is None:
    raise RuntimeError("crs=None")

if template_crs is None:
    raise RuntimeError("terrain.tif no CRScrs=None")

if gdf.crs != template_crs:
    gdf = gdf.to_crs(template_crs)
    print("[OK] reprojected gdf to:", template_crs)

geoms = [geom for geom in gdf.geometry if geom is not None and not geom.is_empty]
if len(geoms) == 0:
    raise RuntimeError("geometry empty")

parks_mask = rasterize(
    [(geom, 1) for geom in geoms],     
    out_shape=template_shape,
    transform=template_transform,
    fill=0,                           
    dtype="uint8",
    all_touched=True                  
)

parks_suit = np.where(parks_mask == 1, PARK_VALUE, NONPARK_VALUE).astype("float32")

if template_nodata is not None:
    parks_suit = np.where(terrain_arr == template_nodata, NODATA, parks_suit)

with rasterio.open(TERRAIN_TIF) as src:
    profile = src.profile.copy()

profile.update(
    dtype="float32",
    count=1,
    nodata=NODATA,
    compress="deflate"
)

with rasterio.open(OUT_TIF, "w", **profile) as dst:
    dst.write(parks_suit, 1)

print(f"[OK] Saved constraint raster: {OUT_TIF}")

arr_plot = np.where(parks_suit == NODATA, np.nan, parks_suit)

plt.figure(figsize=(6, 8))
im = plt.imshow(arr_plot, extent=template_extent, origin="upper", vmin=0, vmax=1)
plt.title("Protected Areas Constraint (0=excluded, 1=allowed)", fontsize=12)
cbar = plt.colorbar(im, fraction=0.035, pad=0.02)
cbar.set_label("Constraint mask", fontsize=10)
plt.xlabel("Easting (m)")
plt.ylabel("Northing (m)")
plt.tight_layout()
plt.savefig(OUT_PNG, dpi=300, bbox_inches="tight")
plt.close()

print(f"[OK] Saved figure: {OUT_PNG}")
```
